# Business Understanding
Untuk memahami bagaimana channel YouTube teratas berperforma, kita akan fokus pada:

- Subscriber Count – Untuk melihat popularitas channel.
- Total Views – Untuk memahami seberapa banyak konten mereka dikonsumsi.
- Genre Distribution – Untuk melihat kategori konten yang dominan.
- Rata-rata Views per Video – Untuk mengetahui engagement per video.

# IMPORT LIBRARY

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# MEMBACA DATA

```python
df = pd.read_excel("/content/top-300-youtube-channels.xlsx")
```

BAR CHART TOP 10 CHANNEL YOUTUBE BERDASARAKAN SUBSCRIBER

```python
top_10_channels = df.nlargest(10, "Subscriber_Count")

plt.figure(figsize=(25, 12))
bars = plt.barh(top_10_channels["Channel_Name"], top_10_channels["Subscriber_Count"], color="red")
plt.xlabel("Jumlah Subscriber")
plt.ylabel("Nama Channel")
plt.title("10 Channel YouTube Teratas berdasarkan Subscriber")
plt.gca().invert_yaxis()

for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2,
             f"{int(bar.get_width()):,}", va='center', ha='left', fontsize=10)

plt.show()
```

PIE CHART DISTRIBUSI GENRE CHANNEL YOUTUBE

```python
genre_counts = df["Genre"].value_counts()

plt.figure(figsize=(16, 13))
plt.pie(genre_counts, labels=genre_counts.index, autopct="%1.1f%%", colors=['lightcoral', 'lightblue', 'lightgreen', 'gold', 'red', 'blue', 'green', 'yellow', 'coral', 'pink', 'purple'])
plt.title("Distribusi Genre Channel YouTube")
plt.axis("equal")
plt.show()
```

SCATTER PLOT HUBUNGAN ANTARA SUBSCRIBER DAN TOTAL VIEWS

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["Subscriber_Count"], df["Video_Views"], c='blue', alpha=0.5)
plt.xlabel("Jumlah Subscriber")
plt.ylabel("Total Views")
plt.title("Hubungan antara Subscriber dan Total Views")
plt.xscale("log")
plt.yscale("log")
plt.show()
```

BOX PLOT DISTRIBUSI RATA-RATA VIEWS PER VIDEO BERDASARKAN GENRE

```python
plt.figure(figsize=(12, 6))
df.boxplot(column="Rata_Rata_Views_per_Video", by="Genre", vert=False, patch_artist=True)
plt.xlabel("Rata-rata Views per Video")
plt.ylabel("Genre")
plt.title("Distribusi Rata-rata Views per Video Berdasarkan Genre")
plt.suptitle("")
plt.xscale("log")
plt.show()
```

# INSIGHT
1. Subscriber Tidak Selalu Memiliki Views Tinggi
- Tidak semua channel dengan banyak subscriber memiliki views tinggi, menandakan engagement yang bervariasi.
2. Genre Dominan
- Beberapa genre memiliki lebih banyak channel, menunjukkan persaingan tinggi di kategori tersebut.
3. Outlier di Views per Video
- Beberapa genre memiliki video dengan views ekstrem, menunjukkan adanya potensi konten viral.
4. Top Channel Punya Strategi Unik
- Channel dengan subscriber terbanyak kemungkinan memiliki strategi khusus dalam konten dan promosi.
